```python
punctuation_chars = ["'", '"', ",", ".", "!", ":", ";", '#', '@']
# lists of words to use
positive_words = []
with open("/Users/aryan/Library/Mobile Documents/com~apple~TextEdit/Documents/positive_words.txt") as pos_f:
    for lin in pos_f:
        if lin[0] != ';' and lin[0] != '\n':
            positive_words.append(lin.strip())


negative_words = []
with open("/Users/aryan/Library/Mobile Documents/com~apple~TextEdit/Documents/negative_words.txt") as pos_f:
    for lin in pos_f:
        if lin[0] != ';' and lin[0] != '\n':
            negative_words.append(lin.strip())
            
def strip_punctuation(s):
    punctuation_chars = ["'", '"', ",", ".", "!", ":", ";", '#', '@']
    string = s
    for ch in string:
        if ch in punctuation_chars:
            s=s.replace(ch,"")
    return s


def get_pos(sentence):
    s = sentence.lower()
    s = strip_punctuation(s)
    s = s.split()
    positivenumber = 0
    for word in s:
        if word in positive_words:
            positivenumber +=1
    return positivenumber       

def get_neg(sentence):
    s = sentence.lower()
    s = strip_punctuation(s)
    s = s.split()
    negativenumber = 0
    for word in s:
        if word in negative_words:
            negativenumber +=1
    return negativenumber  

with open('resulting_data.csv','w') as writeto:
    with open('/Users/aryan/Library/Mobile Documents/com~apple~TextEdit/Documents/project_twitter_data.txt') as readfrom:
        data = readfrom.readlines()
        counter = 0
        writeto.write('Number of Retweets, Number of Replies, Positive Score, Negative Score, Net Score\n')
        for d in data:
            s=d
            s = s.strip()
            s = s.split(',')
            
            retweets = s[1]            
            replies = s[2]
            positivescore = get_pos(s[0])
            negativescore = get_neg(s[0])
            if positivescore >= negativescore:
                netscore = positivescore - negativescore
            else:
                netscore = -(negativescore - positivescore)
            print(netscore)   
            print(str(retweets)+','+str(replies)+','+str(positivescore)+','+str(negativescore)+','+str(netscore))
            if counter !=0:
                writeto.write(str(retweets)+','+str(replies)+','+str(positivescore)+','+str(negativescore)+','+str(netscore)+'\n')
            counter = counter + 1
            
        
```
